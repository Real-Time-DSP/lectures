```julia
using WebIO;
using Mux;
using Interact;
using Plots;
```

```julia
g(t,μ,σ,ω₀) = √(2/(σ*√pi)) .* exp.((-(t.-μ).^2)/(2*σ^2)) .* cos.(ω₀.*t);
G(f,μ,σ,ω₀) = g(f,ω₀,1/σ,0) + g(f,-ω₀,1/σ,0);
s(f,μ,σ,ω₀,n) = g(f,0,1/σ,0) + sum([ G(f,μ,σ,k*ω₀) for k=1:n])
S(f,μ,σ,ω₀,n) = sum([ g(f,μ,σ,k*ω₀) for k=0:n])
t = range(-10,10,1000);
f = range(-10,10,1000);
```

```julia
@manipulate for σ in range(0.1,10,101)
    p1 = plot(t,g(t,0,σ,0),xlab="time",label=nothing);
    p2 = plot(f,g(f,0,1/σ,0),xlab="frequency",label=nothing);
    plot(p1,p2,layout=(2,1))
end
```

```julia
@manipulate for σ in range(0.1,10,101), ω₀ in range(0,10,101)
    p1 = plot(t,G(t,0,σ,ω₀),xlab="time",label=nothing);
    p2 = plot(f,g(f,0,σ,ω₀),xlab="frequency",label=nothing);
    plot(p1,p2,layout=(2,1))
end
```

```julia
@manipulate for σ in range(0.1,10,101), ω₀ in range(0,10,101), n in 1:10
    p1 = plot(t,s(t,0,σ,ω₀,n),xlab="time",label=nothing);
    p2 = plot(f,S(f,0,σ,ω₀,n),xlab="frequency",label=nothing);
    plot(p1,p2,layout=(2,1))
end
```

[http://localhost:9999](http://localhost:9999)

```julia
function app(request)
    @manipulate for σ in range(0.1,10,101), ω₀ in range(0,10,101), n in 1:10
        p1 = plot(t,s(t,0,σ,ω₀,n),xlab="time",label=nothing);
        p2 = plot(f,S(f,0,σ,ω₀,n),xlab="frequency",label=nothing);
        plot(p1,p2,layout=(2,1))
    end
end
fetch(WebIO.webio_serve(page("/",app),9999))
```
